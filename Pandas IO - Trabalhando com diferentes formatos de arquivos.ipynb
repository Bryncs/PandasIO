# Manipulando arquivos JSON

## Lendo arquivos JSON

```python
path_pacientes = f'{path}pacientes.json'
```

```python
dados_pacientes = pd.read_json(path_pacientes)
```

```python
dados_pacientes
```

```python
path_pacientes2 = f'{path}pacientes_2.json'
```

```python
dados_pacientes_2 = pd.read_json(path_pacientes2)
```

```python
dados_pacientes_2
```

## Normalizando arquivos JSON

```python
df_normalizado = pd.json_normalize(dados_pacientes_2['Pacientes'])
```

```python
df_normalizado
```

## Escrevendo arquivos JSON

```python
df_normalizado.to_json('dados/historico_pacientes_normalizado.json')
```
